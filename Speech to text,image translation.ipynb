```python
from pydub import AudioSegment
mp3_file = ('/content/WhatsApp Audio 2025-05-08 at 22.47.49_4f014683.waptt (online-audio-converter.com).mp3')
wav_file = ('/content/WhatsApp Audio 2025-05-08 at 22.47.49_4f014683.waptt (online-audio-converter.com).wav')
```

```python
audio = AudioSegment.from_mp3(mp3_file)
audio.export(wav_file, format='wav')
```

```python
import speech_recognition as sr
reco = sr.Recognizer()
with sr.AudioFile(wav_file) as source:
  audio = reco.record(source)
  text = reco.recognize_google(audio)
  print(text)
```
