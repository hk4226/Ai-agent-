```python
def count_primes_simple(numbers):
    """Counts the number of prime numbers in a set.  Suitable for smaller sets."""
    count = 0
    for num in numbers:
        if num > 1:  # Prime numbers are greater than 1
            is_prime = True
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    is_prime = False
                    break
            if is_prime:
                count += 1
    return count

# Get input from the user
input_str = input("Enter a set of numbers separated by spaces: ")
try:
    numbers = {int(x) for x in input_str.split()}  #Convert input string to a set of integers.
    prime_count = count_primes_simple(numbers)
    print(f"The number of prime numbers in the set is: {prime_count}")
except ValueError:
    print("Invalid input. Please enter numbers separated by spaces.")
```
